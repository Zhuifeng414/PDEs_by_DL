```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from model import NeuralNet
```

```python
model = torch.load('model.pth')
```

```python
x = torch.arange(-2.2, 2.2, 0.01).view(-1, 1)
x_np = x.numpy()
```

```python
PATH = 'model.pth'
model = NeuralNet(1, 20, 1)
model.load_state_dict(torch.load(PATH))
```

```python
a, b, sigma = 0.3, 0.5, 0.5
```

```python
p = lambda x: np.exp((2*a*x**2 - b*x**4) / (2*(sigma**2)))
```

```python
import scipy.integrate as integrate

result = integrate.quad(p, -2.2, 2.2)[0]
```

```python
y_nn = model(x)
y_real = p(x_np) / result
```

```python
plt.plot(y_nn.detach().numpy(), label='neural_network')
plt.plot(y_real, label = 'real')
plt.legend()
```
